```python
'''Generates figure 4 from the paper.'''

import numpy as np
import matplotlib.pyplot as plt
#import csv 
data = np.loadtxt('RMSE.csv', delimiter=',')
indices = [1] + [i for i in range(7,7*14,7)]
indices 
```

```python
#boxplot, do not show  outliers
plt.boxplot(data.T, showfliers=False)
#use the indices to set the xticks
plt.xticks(range(1, len(indices)+1), indices)
#set the x-axis label
plt.xlabel('decorrelation time $1/\lambda$')
#set the title
plt.title('RMSE of $\\beta(t)$')
#svae the figure
plt.savefig('fig_RMSE_lambda_all.png',dpi=300)
```
