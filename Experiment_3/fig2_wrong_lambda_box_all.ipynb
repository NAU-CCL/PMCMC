```python
''''Generates figure 7 and figure 8 from the paper. The left two plots of figure 5 are a subset of this figure, where lambda equals 14 and 63.'''

import sys
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.backends.backend_pdf import PdfPages

sys.path.append('../')

n_max=15

# indices = 1, 7, 14, ..., 98
indices = [i for i in range(0,7*n_max,7)]
indices[0] = 1
run=10

    
```

```python
#make two boxplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
lam_i = 14
data = np.genfromtxt('RMSE_trueT_'+str(lam_i)+'.csv', delimiter=' ')
ax[0].boxplot(data.T, positions=indices, widths=5, showfliers=False)
ax[0].set_title('RMSE of $\\beta(t)$, true $1/\lambda = $'+str(lam_i))
ax[0].set_xlabel('$1/\lambda$ used in PF')
ax[0].set_xticklabels(indices, rotation=45)
#rotate the x-axis labels


lam_i = 63
data = np.genfromtxt('RMSE_trueT_'+str(lam_i)+'.csv', delimiter=' ')
ax[1].boxplot(data.T, positions=indices, widths=5, showfliers=False)
ax[1].set_title('RMSE of $\\beta(t)$, true $1/\lambda = $'+str(lam_i))
ax[1].set_xlabel('$1/\lambda$ used in PF')
ax[1].set_xticklabels(indices, rotation=45)

plt.xticks(rotation=45)
#save the plot as png
plt.savefig('boxplot_RMSE_two_examples.png', dpi=300)
```

```python
#make a 3x5 plot of the 15 lambda values
fig, ax = plt.subplots(3,5, figsize=(15,8))

for i,lam_i in enumerate(indices):

    data = np.genfromtxt('WIS_trueT_'+str(lam_i)+'.csv', delimiter=' ')
    #make the boxplot with xticks being indices array, each column  is a different lambda value
    ax[i//5,i%5].boxplot(data.T, widths=0.6, showfliers=False)
    ax[i//5,i%5].set_title('WIS for $\\beta(t)$ with true $\lambda$ = 1/'+str(lam_i))

    #x ticks are squeezed together, so rotate them
    ax[i//5,i%5].set_xticklabels(indices, rotation=45)
    #set x labels
    ax[i//5,i%5].set_xlabel('$1/\lambda$ set to PF')

#save the plot to a png file with dpi 300
plt.tight_layout()
plt.savefig('boxplot_wis_all.png', dpi=300)

```

```python
#make a 3x5 plot of the 15 lambda values
fig, ax = plt.subplots(3,5, figsize=(15,8))

for i,lam_i in enumerate(indices):

    data = np.genfromtxt('RMSE_trueT_'+str(lam_i)+'.csv', delimiter=' ')
    #make the boxplot with xticks being indices array, each column  is a different lambda value
    ax[i//5,i%5].boxplot(data.T, widths=0.6, showfliers=False)
    ax[i//5,i%5].set_title('RMSE for $\\beta(t)$ with true $1/\lambda$ = '+str(lam_i))

    #x ticks are squeezed together, so rotate them
    ax[i//5,i%5].set_xticklabels(indices, rotation=45)
    #set x labels
    ax[i//5,i%5].set_xlabel('$1/\lambda$ used in PF')

#save the plot to a png file with dpi 300
plt.tight_layout()
plt.savefig('boxplot_rmse_all_time.png', dpi=300)
```
